```python
"""
                                        Summary
The code in the evaluation file is used to compress the the data set video using XVID and H.264 codecs and then calculate its PSNR and SSIM
value and also note down its file size value on the hard disk.
For running this code the required dependencies are;
 1. Pip install numpy       Used for matrices handling
 2. Pip install opencv-python       Used for processing frames/videos
 3. Pip install scikit-image        Used for PSNR and SSIM value calculations
"""


import numpy as np
import cv2 as cv
```

```python
#This is a user define function to calculte the PSNR and SSIM values of the frames. It take two frames (reference_frame,decoded_frame) in the
# input arguments to calculate the above mentioned values. 
#This function uses open-cv and "scikit images" library for SSIM and PSNR value calculation.

def eval_matric(reference_frame,decoded_frame):
    import cv2 as cv
    from skimage.metrics import peak_signal_noise_ratio, structural_similarity

    # Convert frames to grayscale (assuming frames are color images)
    gray_frame1_reference = cv.cvtColor(reference_frame, cv.COLOR_BGR2GRAY)
    gray_frame2_decoded = cv.cvtColor(decoded_frame, cv.COLOR_BGR2GRAY)

    # Calculate PSNR
    psnr_value = peak_signal_noise_ratio(gray_frame1_reference, gray_frame2_decoded)

    # Calculate SSIM
    ssim_value, _ = structural_similarity(gray_frame1_reference, gray_frame2_decoded, full=True)

   
    return psnr_value, ssim_value
```

```python
# # Encoding video file using XVID, H.264(HEVC) codecs
# The below code is used to extract frames from the data set videos and encode it using XVID/H.264 codecs

video = cv.VideoCapture('s4.avi')
ret, frame = video.read()
height, width, channels = frame.shape
count=0
v_enc = cv.VideoWriter('enc_s4_eval_264.mp4', cv.VideoWriter_fourcc(*'x264'), video.get(cv.CAP_PROP_FPS), (width, height))    # For H.264 codec
# v_enc = cv.VideoWriter('enc_s4_eval_xvid.mp4', cv.VideoWriter_fourcc(*'xvid'), video.get(cv.CAP_PROP_FPS), (width, height)) # For XVID codec
v_enc.set(cv.VIDEOWRITER_PROP_QUALITY, 100)

v_enc.write(frame)
count=1
print("frame count",count)

# while (video.isOpened()):  # To process whole file
while (count):  # To process 200 frames

    ret, frame_C = video.read()
    
    if (count<=200):
        count+=1
        
        v_enc.write(frame_C)
        print("frame count",count)
    else:
        print("Process completed")
        video.release()
        v_enc.release()
        break
video.release()
v_enc.release()
```

```python
#Code for calculating PSNR and SSIM values for existing CODECs(XVID,H.264)
# The below code is used for calculating the PSNR and SSIM value of the XVID nad H.264 encode video against the original video

# PSNR and SSIM for source video vs existing codec (e.g h.264, xvid) encoded video
psnr_val_exist=[]
ssim_val_exist=[]
enc_video = cv.VideoCapture('enc_s2_eval_264.mp4') # Loading video encodeb by h.264, xvid 
ref_video=cv.VideoCapture('s2.avi')  #Loading  Original video 

for i in range (200): # As the encoded videos caontains 200 frames so the loop is set for 200 iterations
    
    ret, enc_frame = enc_video.read() # Ectarcting frames from the encoded video
       
    ret1, reference_frame = ref_video.read() # Extracting frame from the original video
    # Calculating the SSIM and PSNR value and then storing it in their correspoing arrays for average value calculation
    psnr_value,ssim_value=eval_matric(reference_frame,enc_frame)  
    print(f'PSNR: {psnr_value:.2f} dB')
    print(f'SSIM: {ssim_value:.4f}')
    psnr_val_exist.append(psnr_value)
    ssim_val_exist.append(ssim_value)
    
# Closing the videos files and displaying the results
enc_video.release()
ref_video.release()
print('processing finished')
print('Existing algorithm')
print(f'SSIN ={np.mean(ssim_val_exist):.4f}','\n',f'PSNR ={np.mean(psnr_val_exist):.2f} dB')
```
